# 11- Tic_tac_toe
One hot encoding

```python
import pandas as pd
import numpy as np
from pathlib import Path

working_directory = Path.cwd()

tic_tac_toe_directory = working_directory / '11-tic+tac+toe+endgame' / 'tic-tac-toe.data'
tic_tac_toe_directory
```

```python
df_tic_tac_toe = pd.read_csv(tic_tac_toe_directory, header=None)
df_tic_tac_toe
```

```python
tic_tac_toe_columns = ['top-left-square','top-middle-square', 'top-right-square', 'middle-left-square','middle-middle-square', 'middle-right-square', ' bottom-left-square', 'bottom-middle-square', 'bottom-right-square', 'Target']
df_tic_tac_toe.columns = tic_tac_toe_columns
df_tic_tac_toe['Target'] = df_tic_tac_toe['Target'].map({'positive':1, 'negative':0})
df_tic_tac_toe['Target'].value_counts()
```

```python
df_tic_tac_toe.isnull().sum() #--> No missing values
```

```python
# Perform one-hot encoding
df_tic_tac_toe_encoded = pd.get_dummies(df_tic_tac_toe.iloc[:,:-1]) # --> except the last columns
df_tic_tac_toe_encoded = pd.concat([df_tic_tac_toe_encoded, df_tic_tac_toe.iloc[:,-1]], axis=1)
df_tic_tac_toe_encoded
```

```python
# Convert True and False to 1 and 0 respectively
df_tic_tac_toe_encoded = df_tic_tac_toe_encoded.astype(int)
df_tic_tac_toe_encoded
```

```python
# Save the DataFrame to a CSV file
csv_filename = '11-tic+tac+toe.csv'
df_tic_tac_toe_encoded.to_csv(csv_filename, index=False)  # Set index=False to exclude the index column
```

# 6- Heart

```python
import pandas as pd
import numpy as np
from pathlib import Path

working_directory = Path.cwd()

heart_directory = working_directory / '6-statlog+heart' / 'statlog_heart.csv'
heart_directory
```

```python
df_heart = pd.read_csv(heart_directory).drop('patient_id',axis=1)
df_heart 
```

```python
heart_target_col =  df_heart.loc[:,'class']
df_heart = df_heart.drop('class',axis=1)

categorical_columns = ['sex', 'chest pain type', 'fasting blood sugar',
       'resting electrocardiographic results', 'exercise induced angina','thal']

num_columns = [c for c in list(df_heart.columns) if c not in categorical_columns]
```

```python
df_heart_encoded = pd.get_dummies(df_heart.loc[:,categorical_columns].astype('category')).astype(int)
df_heart_encoded = pd.concat([df_heart_encoded, df_heart.loc[:,num_columns]], axis=1)
df_heart_encoded['Target'] = heart_target_col
df_heart_encoded
```

```python
# Save the DataFrame to a CSV file
csv_filename = '6-heart.csv'
df_heart_encoded.to_csv(csv_filename, index=False)  # Set index=False to exclude the index column
```

```python
df_cancer.info()
```

```python
df_cancer.isnull().sum()
```

```python
df_cancer = df_cancer.replace('?', np.nan)
```

```python
df_cancer.isnull().sum()
```

```python
df_cancer['Bare Nuclei'] = df_cancer['Bare Nuclei'].fillna(df_cancer['Bare Nuclei'].mode().iloc[0])
```
